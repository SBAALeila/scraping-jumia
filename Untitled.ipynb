```python
from bs4 import BeautifulSoup 
import requests
import pandas as pd
import numpy.ma.mrecords as mrecords
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url = 'https://www.jumia.ma/sante-beaute-beaute-soins-personnels/?page='
```

```python
req = requests.get(url)
```

```python
status = req.status_code
encoding = req.encoding
html_doc = req.text
status
```

```python
html_doc
```

```python
soup = BeautifulSoup(html_doc, 'html.parser')
```

```python
soup
```

```python
columns = {'name':[] , 'price':[], 'img url': [], 'brand': [], 'livraison': [], "review": []}
```

```python
for page in range(1,10):
    print('---', page, '---')
    r = requests.get(url + str(page))
    soup = BeautifulSoup(r.content, "html.parser")
    data = soup.find('div',{'class' : '-paxs row _no-g _4cl-3cm-shs'}).find_all('article' , {'class' : 'prd _fb col c-prd'})
    for item in data:
        img=item.find('a').find('div' , {'class' : 'img-c'}).find('img',{'class':'img'})

        name = item.find('a').find('div' , {'class' : 'info'}).find('h3' , {'class' : 'name'})
    
        
        price = item.find('a').find('div' , {'class' : 'info'}).find('div' , {'class' : 'prc'})
        
        brand= item.find('a').get('data-brand')
        
        livraison = item.find('a').find('div' , {'class' : 'info'}).find('p' , {'class' : 'shipp'})
        
        review= item.find('a',{'class' : 'core'}).get('data-dimension27')
       
        
        columns['name'].append(name.text)
        columns['price'].append(price.text)
        columns['img url'].append(img.get('data-src'))
        columns['brand'].append(brand)
        if livraison is not None:
             columns['livraison'].append(livraison.text)
        else:
              columns['livraison'].append(None)
        if review is not None:
             columns['review'].append(review)
        else:
             columns['review'].append(None)
        
data = pd.DataFrame(columns)
data.to_excel('data.xlsx')
```
